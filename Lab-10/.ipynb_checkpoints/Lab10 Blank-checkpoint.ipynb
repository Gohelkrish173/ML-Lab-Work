# Step 1: Import Libraries

```python
import pandas as pd
from sklearn.datasets import make_blobs,make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram,linkage
import numpy as np
```

# Step 2: Generate dummy data using make_blobs

```python
x,y = make_blobs(n_samples=30,centers=3,random_state=42,cluster_std=1)
x = StandardScaler().fit_transform(x)
```

# Step 3: Plot Generated data 

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1], s=30,cmap='virdis',c="blue")
plt.title("Raw Data")
```

# Step 4: Apply Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
model.linkage
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1], s=30,c=model.labels_,)
plt.title("Raw Data")
```

# Step 5: Display Dendrogram

```python
linked = linkage(x, method="ward")
plt.figure(figsize=(10,5))
dendrogram(linked)
plt.show()
```

# Step 6: Generate dummy data using make_moons

```python
x,y = make_moons(n_samples=200,noise=0.1,random_state=50)
x = StandardScaler().fit_transform(x)
```

# Step 7: Plot Generated data

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],s = 30 ,cmap='viridis')
plt.title("Raw Data")
plt.show()
```

# Step 8: apply DBSCAN Clustering 

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.2,min_samples=5).fit(x)
```

```python
model.fit_predict(x)
```

```python
model.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=model.labels_,cmap='rainbow',edgecolor="red")
```

# Step 9: apply K Mean Clustering on same data

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=3,random_state=42,n_init=10)
```

```python
kmean.fit_predict(x)
```

```python
plt.figure(figsize=(6,6))

centerdata = kmean.cluster_centers_

plt.scatter(x[:,0],x[:,1],c=kmean.labels_)
plt.scatter(centerdata[:,0],centerdata[:,1],marker="x")
plt.show()
```

# Mall Customers Segmentation using DBSCAN

### Perform Following steps on Dataset for  customer segmentation
-> Load & Explore the Dataset <br>
-> Data Preprocessing <br>
-> Apply DBSCAN <br>
-> Visualizing Customer Segments <br>
-> Outliers (high-spending VIP customers or unusual shoppers <br>

```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
x = data.iloc[:,[3,4]].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
```

```python
newdbscan = dbscan.fit_predict(x)
newdbscan
```

```python
data['clusters'] = newdbscan
```

```python
x1 = []
x2= []
x3 = []

for i,j in zip(x,newdbscan):
    if(j == 0):
        x2.append(i)
    elif j == 1:
        x3.append(i)
    elif j == -1 :
        x1.append(i)
        
x1 = StandardScaler().fit_transform(x1)
x2 = StandardScaler().fit_transform(x2)
x3 = StandardScaler().fit_transform(x3)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x1[:,0],x1[:,1],label="-1")
plt.scatter(x2[:,0],x2[:,1],label="0")
plt.scatter(x3[:,0],x3[:,1],label="1")
plt.legend(title="clusters")
plt.title("Row Data")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x[:,0],x[:,1],c=newdbscan,cmap="viridis",s=10)
plt.show()
```
