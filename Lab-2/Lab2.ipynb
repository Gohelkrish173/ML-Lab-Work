## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2. Import the dataset

```python
iphone_data = pd.read_csv("iphones.csv")
iphone_data
plt.scatter(iphone_data["iphonenumber"],iphone_data["price"])
```

## Step 3. Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 4. Create LinearRegression Model

```python
model = LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
X = iphone_data['iphonenumber'].values.reshape(-1,1)
Y = iphone_data['price']
```

## Step 6. Fit the Model (Training)

```python
model.fit(X,Y)
```

## Step 7. Prdict the value

```python
N = input("Enter Iphone Number : ")

model.predict([[int(N)]])
```

```python
newY = model.predict(X)
newY
```

## Step 8. Result Visualization

```python
plt.plot(X,Y)
plt.plot(X,newY)
```

```python
plt.scatter(X,Y)
plt.scatter(X,newY,)
```

## Dataset - 2 - Salary Dataset

## Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Step 2.  Read the Data

```python
salary_df = pd.read_csv("Salary_Data.csv")
salary_df
```

## Step 3. View Data using Scatter plot 

```python
plt.scatter(salary_df['YearsExperience'],salary_df['Salary'])
```

## Step 4. Splitting dataset in to input and output

```python
X = salary_df['YearsExperience'].values.reshape(-1,1)
Y = salary_df['Salary']
```

## Step 4. Splitting dataset in to input and output

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2) # every time change train array
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) # every time train array not change
```

```python
x_test
```

## Step 6. Create LinearRegression Model

```python
Model = LinearRegression()
```

## Step 7. Fit the Model (Training)

```python
Model.fit(x_train,y_train)
```

##  Step 6. Predicting the Test set results

```python
N = input("Enter Year of Experience : ")

Model.predict([[int(N)]])
```

##  Step 7. Visualising the Training set results

```python
plt.plot(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
```

##  Step 8. Visualising the Test set results

```python
newY = Model.predict(x_test)
plt.scatter(x_test,y_test)
plt.scatter(x_test,newY)
```
