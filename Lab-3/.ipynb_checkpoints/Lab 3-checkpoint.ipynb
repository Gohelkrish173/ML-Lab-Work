## Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Step 2. Import the dataset

```python
df = pd.read_csv("50_Startups.csv")
df
```

## Step 3 . Check the State Column

```python
df["State"].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
x = df.iloc[::,0:4:]
y = df.iloc[::,4:]
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Import  ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
```

## Step 5.2 .  Import  OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

## Step 5.3 .  Perform Transformation

```python
ct = ColumnTransformer(transformers=[('encoder',
OneHotEncoder(),[3])],remainder="passthrough")
```

```python
X = np.array(ct.fit_transform(x))
X
```

```python
ct.fit_transform(x) # gives this type of value when we can't get .values where split the input and output varialbes
```
